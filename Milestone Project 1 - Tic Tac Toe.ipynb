```python
#intro statemt 

def ttt_intro():
    
    player1 = 'wrong'
    player2 = 'wrong'
    
    #intro message
    print("Welcome  to Tic Tac Toe!") 
    
    #keeps loping until Playeer1 makes a valid choice (X or O)
    while player1 not in ["X", "O"]:

        player1 = input("Player 1: Do you want to be 'X' or 'O' ?")
        
        if player1 not in ["X","O"]:
            print("Sorry, please select 'X' or 'O' ")

    #assigns player 2 to the other option        
    if player1 == "X":
        player2 = "O"
        
    else:
        player2 = "X"

    
    return(player1,player2)


    
```

```python
# what the current board looks like
board= ['#',' ',' ',' ',' ',' ',' ',' ',' ',' ']

def display_board(board):
    
    print('\n')
    print(board[7] + '|' + board[8] + '|' + board[9])
    print('-----')
    print(board[4] + '|' + board[5] + '|' + board[6])
    print('-----')
    print(board[1] + '|' + board[2] + '|' + board[3])
    print('\n')

    
```

```python
#asking Player 1 for their choice

def player1_placement(player1_choice):
    
    #ensures player1's choice is withing the board
    while player1_choice not in ['1','2','3','4','5','6','7','8','9']:
        
        player1_choice = input("Player 1: Please place your marker in a position: ")
        
        if player1_choice not in ['1','2','3','4','5','6','7','8','9']:
            print("Please pick a valid position!")
        
    return int(player1_choice)
        
```

```python
#asking Player 2 for their choices

def player2_placement(player2_choice):
    
    #ensure player2's choice is within the board
    while player2_choice not in ['1','2','3','4','5','6','7','8','9']:
        
        player2_choice = input("Player 2: Please place your marker in a position: ")
        
        if player2_choice not in ['1','2','3','4','5','6','7','8','9']:
            print("Please pick a valid position!")
    
    return int(player2_choice)
    
```

```python
#ensures player1 has not chosen a spot that already has a marker in it
def player1_empty_check(player1_choice,board):
    
    while board[player1_choice] != ' ':
        
        player1_choice = input("Please chose an empty position: ")
        
        player1_choice = int(player1_choice)
        
        if board[player1_choice] != ' ':
            print("Please pick an empty position!")
            
    return int(player1_choice)
```

```python
#ensures player2 has not chosen a spot that already has a marker in it
def player2_empty_check(player2_choice,board):
    
    while board[player2_choice] != ' ':
        
        player2_choice = input("Please chose an empty position: ")
        
        player2_choice = int(player2_choice)
        
        if board[player2_choice] != ' ':
            print("Please pick an empty position!")
            
    return int(player2_choice)
```

```python
#inserting player 1 choice
def player1_insert(board,player1_choice,player1):
    
    board[player1_choice] = player1
    
    return board
```

```python
# inserting player 2 chocie
def player2_insert(board,player2_choice,player2):
    
    board[player2_choice] = player2
    
    return board
```

```python
#checking if player 1 has won
def p1_winner_check(board,player1):
    
    player1_declare = 'no'
    
    if board[1] == player1 and board[2] == player1 and board[3] == player1: #p1 bottom row check
        player1_declare = 'yes'
        
    elif board[4] == player1 and board[5] == player1 and board[6] == player1: #p1 middle row check
        player1_declare = 'yes'
        
    elif board[7] == player1 and board[8] == player1 and board[9] == player1: #p1 top row check
        player1_declare = 'yes'
        
    elif board[1] == player1 and board[4] == player1 and board[7] == player1: #p1 left collumn check
        player1_declare = 'yes'
        
    elif board[2] == player1 and board[5] == player1 and board[8] == player1: #p1 middle column check
        player1_declare = 'yes'
        
    elif board[3] == player1 and board[6] == player1 and board[9] == player1: #p1 right column check
        player1_declare = 'yes'
        
    elif board[1] == player1 and board[5] == player1 and board[9] == player1: #p1 bot left to top right diagonal check
        player1_declare = 'yes'
        
    elif board[3] == player1 and board[5] == player1 and board[7] == player1: #p1 bot right to top left diagonal check
        player1_declare = 'yes'
        
    return player1_declare
```

```python
#checking if player 2 has won
def p2_winner_check(board,player2):
    
    player2_declare = 'no'
    
    if board[1] == player2 and board[2] == player2 and board[3] == player2: #p2 bottom row check
        player2_declare = 'yes'
        
    elif board[4] == player2 and board[5] == player2 and board[6] == player2: #p2 middle row check
        player2_declare = 'yes'
        
    elif board[7] == player2 and board[8] == player2 and board[9] == player2: #p2 top row check
        player2_declare = 'yes'
        
    elif board[1] == player2 and board[4] == player2 and board[7] == player2: #p2 left column check
        player2_declare = 'yes'
        
    elif board[2] == player2 and board[4] == player2 and board[8] == player2: #p2 middle column check
        player2_declare = 'yes'
        
    elif board[3] == player2 and board[6] == player2 and board[9] == player2: #p2 right column check
        player2_declare = 'yes'
        
    elif board[1] == player2 and board[5] == player2 and board[9] == player2: #p2 bot left to top right diagonal check
        player2_declare = 'yes'
        
    elif board[3] == player2 and board[5] == player2 and board[7] == player2: #p2 bot right to top left diagonal check
        player2_declare = 'yes'
        
        
    return player2_declare
```

```python
#check if the bard is full
def board_full_check(board):
    
    board_full = 'no'
    
    if board[1] == ' ':
        return board_full
    
    elif board[2] == ' ':
        return board_full
        
    elif board[3] == ' ':
        return board_full
        
    elif board[4] == ' ':
        return board_full
        
    elif board[5] == ' ':
        return board_full
        
    elif board[6] == ' ':
        return board_full
        
    elif board[7] == ' ':
        return board_full
        
    elif board[8] == ' ':
        return board_full
        
    elif board[9] == ' ':
        return board_full
        
    else:
        board_full = 'yes'
        return board_full
```

```python
#declaring the winner
def winner_declare(player1_declare, player2_declare):
    
    if player1_declare == 'yes':
    
        print("Congradulations Player 1, you are the winner!")
              
    elif player2_declare == 'yes':
        print("Congradulations Player 2, you are the winner!")
        
    else:
        print("It's a tie game!")
```

```python
#prints 100 blank lines to clear screen
def clear_screen():
    print('\n'*100)
```

```python
def play_again(another_round):
    
    while another_round == 'Yes':
        
        another_round = input('Would you like to play again? Yes or No: ')
        
        if another_round not in ['Yes', 'No']:
            print("Please pick a valid answer!")
    
    return another_round
```
